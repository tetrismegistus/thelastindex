# rates

```python
material_prices = [24.34, 53.04]
commission_fee = 1.6;
shipping_rates = [0, 22.50, 52.50, 79]

for material_price in material_prices:
    for shipping_rate in shipping_rates:
        print(f"{material_price} * {commission_fee} + {shipping_rate} = {material_price * commission_fee + shipping_rate}")
```

# commission fee

```python
for material_price in material_prices:
    for shipping_rate in shipping_rates:
        print(f"commission fee: {(material_price * commission_fee + shipping_rate) - (material_price + shipping_rate)}")
```
